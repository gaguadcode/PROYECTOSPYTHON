```python
import pandas as pd
import missingno as msno

# cargo los datos a dataframe
file_path = 'avgIQpercountry.csv'
df = pd.read_csv(file_path)

print(df.head())
```

```python
msno.matrix(df)
```

```python
# diagrama de barras de dartos faltantes
msno.bar(df)
```

```python
# Display a heatmap of missing values
msno.heatmap(df)
```

```python


# Aplica la mediana usando una función lambda solo a columnas numéricas
df_filled = df.apply(lambda col: col.fillna(col.median()) if pd.api.types.is_numeric_dtype(col) else col)

# Ahora, df_filled contiene el DataFrame original con los valores faltantes reemplazados por la mediana solo en columnas numéricas.
msno.bar(df_filled)
```
