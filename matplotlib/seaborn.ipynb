```python
import seaborn as sns
import matplotlib.pyplot as plt

# cargo el datset de exoplanetas de seaborn
planets = sns.load_dataset("planets")
planets.head()

# busco correlacion entre periodo orbital y la masa
sns.scatterplot(x="orbital_period", y="mass", hue="method", data=planets)

# Set plot title and labels
plt.title("Scatter Plot of Orbital Period vs. Mass")
plt.xlabel("Orbital Period (days)")
plt.ylabel("Mass (M_jupiter)")  # masa dada en terminos de jupiter

```

```python
sns.barplot(x="mass", y="method", data=planets)


plt.title("Bar Plot of Planet Mass by Discovery Method")
plt.xlabel("Mass (M_jupiter)")  
plt.ylabel("Discovery Method")

# Show the plot
plt.show()
```

```python

sns.histplot(planets["mass"], bins=30, kde=True)


plt.title("Histogram of Planet Mass")
plt.xlabel("Mass (M_jupiter)")  
plt.ylabel("Frequency")
```
