```python
import pandas as pd
df = pd.read_csv('https://drive.google.com/uc?export=download&id=1033_pgXO86pl9hwlIvGZZtNgmbiYJ7ps')
print(df.to_string(index=False))
```

```python
import pandas as pd

# Parte 1: Exploración Básica del DataFrame

# Mostrar las primeras y últimas 5 filas del DataFrame
print("Primeras 5 filas:")
print(df.head())

print("\nÚltimas 5 filas:")
print(df.tail())

# Obtener información general del DataFrame
print("\nInformación del DataFrame:")
print(df.info())

# Parte 3: Cálculos y Manipulación

# En este caso, puedes explorar relaciones entre género, rentabilidad y edad del protagonista al hacer la película
# Crear una nueva columna 'Ganancias por Minuto'
df['Ganancias por Minuto'] = df['Ganancias'] / df['Duración (minutos)']
# Ejemplo 1: Relación entre género y rentabilidad
relacion_genero_rentabilidad = df.groupby('Género')['Recaudación'].mean()
print("\nRelación entre género y recaudación promedio:")
print(relacion_genero_rentabilidad)

# Ejemplo 2: Relación entre año de emisión y calificación IMDB
relacion_anio_calificacion = df.groupby('Año de Emisión')['Calificación IMDB'].mean()
print("\nRelación entre año de emisión y calificación IMDB promedio:")
print(relacion_anio_calificacion)

#Correlacion entre ganancias y calificacion
correlacion = df['Ganancias por Minuto'].corr(df['Calificación IMDB'])
print(correlacion)
# Mostrar el resultado
print("Coeficiente de correlación entre Ganancias por Minuto y Calificación IMDB:", correlacion)
# Parte 4: Exportar Resultados
# Guardar los resultados del análisis en un nuevo archivo CSV
df.to_csv("resultados.csv", index=False)  # Asegúrate de cambiar "ruta/del/archivo_resultados.csv" con la ruta correcta
```
