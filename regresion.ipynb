```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Simulación con datos aleatorios
X = 1000*(2 * np.random.rand(100, 1))
y = 4 + 3 * X + 1000*(np.random.randn(100, 1))

#creo el modelo y lo entreno
model = LinearRegression()
model.fit(X, y)

# Predicciones
X_new = 1000*(np.array([[0], [2]]))
y_pred = model.predict(X_new)

# plot de los datos y la recta de regresión
plt.scatter(X, y, alpha=0.7, label='Datos Originales')
plt.plot(X_new, y_pred, 'r-', label='Recta de Regresión', linewidth=2)
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

```

# Predicción del modelo en 1830:

```python
X_pred = np.array([[1830]])
y_pred = model.predict(X_pred)

print("El valor predicho en 1830 será:", y_pred[0, 0])
```
